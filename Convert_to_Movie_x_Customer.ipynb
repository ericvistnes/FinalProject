Create Matrix.txt for ratings (Movie by Customer)

```python
fd = open('matrix.txt', 'w')
```

```python
for mov in range(1,10):
    df = movie[(movie['MovieID'] == mov)]
    print(df)
    movierow = "" 
    lastcust = CID[len(CID) - 1] 
    for cust in CID:
        if (cust != lastcust):
            if (((df['CustomerID'] == cust)).any()):
                movierow += str(df[(df['CustomerID'] == cust)].Rating.item())
            else:
                movierow += "?"
            movierow += ","
        else:
            if (((df['CustomerID'] == cust)).any()):
                movierow += str(df[(df['CustomerID'] == cust)].Rating.item())
            else:
                movierow += "?"
    movierow += "\n"
    fd.write(movierow)
```

```python
fd.close()
```
